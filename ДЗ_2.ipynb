## Problem 1

Let $f(x) = 2\sin x - \cos{2x}$. Write a function `beauty_derivative_plot` that takes a point $x$ and plots the function $f(x)$ and its derivative $f'(x)$ in the range $x \in [-2\pi, 2\pi]$. The function should also plot the tangent line to the function at the point $x$.

Example of output for $x = -3, -2, 1.7$:

```python
beauty_derivative_plot(-3)
```

```python
beauty_derivative_plot(-2)
```
<img src="https://i.ibb.co/F4zDq8m0/tangent-line-2-0.png" width="600"/>

```python
beauty_derivative_plot(1.7)
```
<img src="https://i.ibb.co/TBmLXg2h/tangent-line-1-7.png" width="600"/>

````python
import numpy as np
import matplotlib.pyplot as plt

def beauty_derivative_plot(x0):

    def f(x):
        return 2 * np.sin(x) - np.cos(2 * x)

    def f_prime(x):

        return 2 * np.cos(x) + 2 * np.sin(2 * x)

    x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

    y = f(x)
    y_deriv = f_prime(x)

    y0 = f(x0)
    slope = f_prime(x0)


    x_tangent = np.linspace(x0 - 1, x0 + 1, 100)
    tangent_line = y0 + slope * (x_tangent - x0)


    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=r'$f(x)=2\sin x-\cos2x$', color='blue')
    plt.plot(x, y_deriv, label=r"$f'(x)$", color='green')
    plt.plot(x_tangent, tangent_line, '--', label=f'Касательная в x={x0}', color='red')
    plt.scatter([x0], [y0], color='red', zorder=5)


    plt.title(f'График функции f(x), её производной и касательной в точке x={x0}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

beauty_derivative_plot(-3)
beauty_derivative_plot(-2)
beauty_derivative_plot(1.7)
````

- Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for each year in the period.

- Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Show each histogram in a separate plot. Arrange 3 plots in a row.

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('populations.txt', skiprows=1)

means = np.mean(data, axis=0)
std_devs = np.std(data, axis=0)

labels = ['Hares', 'Lynxes', 'Carrots']
colors = ['skyblue', 'lightgreen', 'salmon']

# Создаем фигуру с тремя графиками
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Строим гистограммы для каждого столбца
for idx, (label, color) in enumerate(zip(labels, colors)):
    axes[idx].hist(data[:, idx + 1], bins=12, color=color, edgecolor='black')  # Увеличено количество корзин
    axes[idx].axvline(means[idx + 1], color='red', linestyle='--', linewidth=2, label=f'Mean: {means[idx + 1]:.2f}')  # Среднее
    axes[idx].set_title(f"{label} Population\nMean: {means[idx + 1]:.2f}, Std: {std_devs[idx + 1]:.2f}")
    axes[idx].set_xlabel("Population")
    axes[idx].set_ylabel("Frequency")
    axes[idx].legend()

plt.tight_layout()

plt.show()
````

- Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

- Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('populations.txt', skiprows=1)

years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]

hares_change = np.diff(hares)
lynxes_change = np.diff(lynxes)

plt.figure(figsize=(10, 6))

plt.plot(years[1:], hares_change, label='Change in Hares Population', color='red', marker='o')
plt.plot(years[1:], lynxes_change, label='Change in Lynxes Population', color='black', marker='x')

plt.title('Change in Population of Hares and Lynxes Over Time')
plt.xlabel('Year')
plt.ylabel('Change in Population')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()


correlation_coefficient = np.corrcoef(hares, lynxes)[0, 1]

print(f"Correlation Coefficient between Hares and Lynxes: {correlation_coefficient:.50f}")
````

- Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value
with the true value and print a ratio of the error to the true value.

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('populations.txt', skiprows=1)

years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]

# Линейная регрессия для прогнозирования популяции зайцев на основе рысей и моркови
X = np.column_stack([lynxes, carrots, np.ones(len(lynxes))])
coeffs = np.linalg.lstsq(X, hares, rcond=None)[0]

# Прогнозируем популяцию зайцев для 1920 года
lynx_1920 = lynxes[years == 1920][0]
carrot_1920 = carrots[years == 1920][0]
hare_1920_pred = np.dot([lynx_1920, carrot_1920, 1], coeffs)

new_years_coef = 1.25
ind = np.arange(len(hares) - 1)
balanced_weights = new_years_coef ** ind
balanced_weights /= balanced_weights.sum()
hare_1920_weighted = np.dot(hares[:-1], balanced_weights)

plt.figure(figsize=(12, 6))


plt.plot(years, hares, marker='s', color='#1f77b4', linestyle='-', label="Фактические данные", markersize=8)
plt.scatter(1920, hare_1920_pred, color='#ff7f0e', label=f"Прогноз (Линейная регрессия): {hare_1920_pred:.2f}", zorder=5, s=100)
plt.scatter(1920, hare_1920_weighted, color='#2ca02c', label=f"Прогноз (Баланс. веса): {hare_1920_weighted:.2f}", zorder=5, s=100)

plt.xlabel("Год", fontsize=14, fontweight='bold', color='#333333')
plt.ylabel("Популяция зайцев", fontsize=14, fontweight='bold', color='#333333')
plt.xticks(np.arange(min(years), max(years) + 1, 2), rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.title("Прогноз популяции зайцев на 1920 год", fontsize=16, fontweight='bold')

plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# Реальная популяция зайцев для 1920 года
hare_1920_true = hares[-1]

# Выводим результаты
print(f"Прогноз популяции зайцев (линейная регрессия) на 1920 год: {hare_1920_pred:.2f}")
print(f"Прогноз популяции зайцев (балансированные веса) на 1920 год: {hare_1920_weighted:.2f}")
print(f"Реальная популяция зайцев в 1920 году: {hare_1920_true}")
print(f"Ошибка прогноза (линейная регрессия): {abs(hare_1920_pred - hare_1920_true) / hare_1920_true:.4f}")
print(f"Ошибка прогноза (балансированные веса): {abs(hare_1920_weighted - hare_1920_true) / hare_1920_true:.4f}")
````
